```python
%matplotlib inline
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pylab
from matplotlib.pyplot import subplots

g = nx.from_pandas_edgelist(df_cop_book, 'Source', 'Target')
sub = nx.node_connected_component(g, 33)

import collections
from bokeh.io import show, output_file
from bokeh.plotting import figure
import seaborn as sns
degree_count=collections.Counter(dict(g.subgraph(sub).degree()).values())

fig, ax = plt.subplots()
plt.bar(degree_count.keys(), degree_count.values(), width=0.8, color='#FFD966')
plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

**CENTRALITY MEASURES**
how easy is it to spread an information in the network. which node is most responsible for maximum spread.

```python
#centrality measure of nodes 
degree_central=nx.degree_centrality(g.subgraph(sub))
plt.bar(range(len(degree_central)),degree_central.values(),width=0.6,color='#F4B183')
fig=plt.gcf()
axes=plt.gca()
fig.set_size_inches(15,10)
axes.set_ylim([0,0.05])
plt.title("centrality of each node")
plt.show()
```

***closeness centrality***

To calculate closeness centrality, one first finds the shortest path between each node in the network and all other nodes. The sum of these distances is then used to calculate the closeness centrality score for each node. Nodes with a higher score are considered to be more central in the network.

```python
close_degree_central=nx.closeness_centrality(g.subgraph(sub))
plt.bar(range(len(close_degree_central)),list(close_degree_central.values()),width=0.9,color='#DFA67B')
fig=plt.gcf()
axes=plt.gca()
fig.set_size_inches(15,8)
plt.title("Closeness Centrality")
plt.show()
```

***Betweenness Centrality***

Betweenness centrality quantifies the number of times a node acts as a bridge along the shortest path between two other nodes.


```python
#between 
between_degree_central=nx.betweenness_centrality(g.subgraph(sub))
plt.bar(range(len(between_degree_central)),list(between_degree_central.values()),width=0.8,color='#B73E3E')
fig=plt.gcf()
axes=plt.gca()
fig.set_size_inches(10,8)
plt.title('Betweeness Centrality')
plt.show()
```

***eigen value centrality***

```python
Eigen_central=nx.eigenvector_centrality_numpy(g.subgraph(sub))
plt.bar(range(len(Eigen_central)),Eigen_central.values(),width=0.8,color='#DD5353')
fig=plt.gcf()
axes=plt.gca()
fig.set_size_inches(15,8)
plt.title('Eigen Vector Centrality')
plt.show()
```
